```python
from bigbird import modeling
from bigbird import utils
import tensorflow as tf
import numpy as np
```

```python
bert_config = {
  # transformer basic configs
  "attention_probs_dropout_prob": 0.1,
  "hidden_act": 'gelu',
  "hidden_dropout_prob": 0.1,
  "hidden_size": 768,
  "initializer_range": 0.02,
  "intermediate_size": 3072,
  "max_position_embeddings": 4096,
  "max_encoder_length": 2048 // 4,
  "max_decoder_length": 512 // 4,
  "num_attention_heads": 12,
  "num_hidden_layers": 12,
  "type_vocab_size": 2,
  "scope": 'bert',
  "use_bias": True,
  "rescale_embedding": False,
  "vocab_model_file": None,
  # sparse mask configs
  "attention_type": "block_sparse",
  "norm_type": 'postnorm',
  "block_size": 64,
  "num_rand_blocks": 3,
  "vocab_size": 32000,
  "beam_size": 1,
  "alpha": 0.0,
  "couple_encoder_decoder": True
}
```

```python
model = modeling.TransformerModel(bert_config)
```

```python
X = tf.compat.v1.placeholder(tf.int32, [None, None])
```

```python
r = model(X, training = False)
```

```python
pred_ids = r[0][2]
```

```python
tf.trainable_variables()
```

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
```

```python
%%time

o = sess.run(pred_ids, feed_dict = {X: [[1] * 512]})
```
